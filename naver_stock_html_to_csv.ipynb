```
import pandas as pd
import requests
import urllib
from bs4 import BeautifulSoup
from glob import glob
import datetime

pd.options.display.max_rows=None
```

```
# 실행 전 html 파일들 업로드 필요
## 네이버 금융 > 업종/테마별 시세 > 원하는 업종/테마 선택 > html 파일 다운로드

today = str(datetime.datetime.now().date()).replace('-', '')
print(today)

df_list = []
for fname in glob('*.html'):
    tmp_df = pd.read_html(fname)[2].dropna(how='all').reset_index(drop=True)
    tmp_df = tmp_df[['종목명', '현재가', '전일비', '등락률', '시가총액', '보통주배당금', '외국인비율', 'PER', 'ROE', 'PBR']]  # html 다운로드 시 선택했던 지표명들
    tmp_df['PER_abs'] = tmp_df['PER'].abs()
    tmp_df['분야'] = fname.split('_ 네이버 금융.html')[0].strip()
    tmp_df.to_csv(fname.replace('.html', f' ({today}).csv'), encoding='cp949', index=False)
    df_list.append(tmp_df)

total_df = pd.concat(df_list, ignore_index=True)
total_df.to_csv(f'stocks_total ({today}).csv', encoding='cp949', index=False)

!zip -r stocks_csv.zip *.csv
```
